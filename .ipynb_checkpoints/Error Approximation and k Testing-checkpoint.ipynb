## Testing k

One of the parameters of the Heaviside Approximation is the value $k$. This value determines how sheer the transition between the two constant regions is. The higher the value of k, the larger the derivative is, and thus, the steeper the function is at the point of transition.

```python
# Import block
import numpy as np
import matplotlib.pyplot as plt
import math
import doctest
%matplotlib inline
```

```python
def original_Heaviside (k, x):
    '''A function defining the original, unscaled Heaviside Approximation
    
    Arguments:
    k: Determines steepness of function
    x: the current location
    
    returns: value of function'''
    
    return 1/(1+np.exp(-2*k*x))

def square_Potential (x):
    '''A function defining the strict square well matching the Heaviside parameters
    
    Arguments:
    x: the current location
    
    returns: value of the function'''
    
    if (x>0):
        return 1
    elif(x==0):
        return 0.5
    else:
        return 0
```

```python
values = np.linspace (-5,5)
k = 1
true_function = []
approximal = []
error = []

for i in range(len(values)):
    approximal.append(original_Heaviside(k,values[i]))
    true_function.append(square_Potential(values[i]))
    error.append(true_function[i]-approximal[i])

plt.plot(values, approximal, label = "Heaviside Approximation")
plt.plot(values,true_function, label = "True Value")
plt.legend(loc='best')
plt.title("True Function Versus Approximation")
plt.show()
```

```python
plt.plot(values,error)
plt.title("Error as f(x)")
plt.show()
```
